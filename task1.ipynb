### Q1: Create a function to convert RGB to Grey scale image. Use numpy, pytorch, or tf but don't direct available functions.
Hint: Use the formula: Gray = 0.299R + 0.587G + 0.114*B for the conversion.

```python
import cv2
import numpy as np
```

```python
def rgb_to_grey(rgb_image_path):
  rgb_image=cv2.imread(rgb_image_path,cv2.IMREAD_COLOR)
  Gray = 0.299*rgb_image[:,:,0]+ 0.587*rgb_image[:,:,1] + 0.114*rgb_image[:,:,2]

  return Gray
print(rgb_to_grey('/jeans.jpg'))
```

### Q2 : Create a function to convolve_image taking image, kernel as input and returned convolved image.


```python
import matplotlib.pyplot as plt
def convolve_image(image_path,kernel):
  image=rgb_to_grey(image_path)
  k=kernel.shape[0]
  rows,cols=image.shape
  output_matrix=np.zeros((rows-k+1,cols-k+1))
  for i in range(1,rows-1):
    for j in range(1,cols-1):
      region=image[i-1:i+2,j-1:j+2]
      output_matrix[i-1,j-1]=np.sum(region*kernel)
  plt.figure=plt.figure(figsize=(10, 5))


  plt.subplot(1, 2, 1)
  plt.title("Original Image")
  plt.imshow(image, cmap='gray')
  plt.axis('off')

  plt.subplot(1, 2, 2)
  plt.title("Processed image")
  plt.imshow(output_matrix, cmap='gray')
  plt.axis('off')
  return plt.show()
```

```python
sobel_vertical=np.array([[[0.111,    0.111,    0.111],
[0.111,    0.111,    0.111],
[0.111,    0.111,    0.111]]])
#type(sobel_vertical)
print("5 nd kernel ",convolve_image('/jeans.jpg',sobel_vertical))
```
