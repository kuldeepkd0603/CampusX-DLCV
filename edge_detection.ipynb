```python
import cv2
import numpy as np
```

```python
image_path="/IMG_20240402_141146.jpg"
image=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
image
```

```python
sobel_vertical=np.array([[[1/9,1/9,1/9],
                         [1/9,1/9,1/9]
                          ,[1/9,1/9,1/9]]])
```

```python
k=sobel_vertical.shape[0]
```

```python
image.shape
```

```python
row,cols=image.shape
row,cols
```

```python
output_matrix=np.zeros(( row-k+1 ,cols-k+1 ))
```

```python
for i in range(1,row-1):
  for j in range(1,cols-1):
    region=image[i-1:i+2,j-1:j+2]
    output_matrix[i-1,j-1]=np.sum(region * sobel_vertical)
```

```python
output_matrix
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure=plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Soheb Conv Edge Output")
plt.imshow(output_matrix, cmap='gray')
plt.axis('off')

plt.show()
```
